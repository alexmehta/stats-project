```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
biased_responses = pd.read_csv('bias.csv')
unbias_responses = pd.read_csv('unbias.csv')
```

```python
biased_responses.head()
```

```python
unbias_responses.head()
```

```python
# make 2 variables, passing period and 8 blocks for the 2 questions. With 2 columns for the bias and unbias question
passing_period_responses = pd.DataFrame(columns=['positive bias', 'unbias'])
passing_period_responses['positive bias'] = biased_responses['Passing Period']
passing_period_responses['unbias'] = unbias_responses['Passing Period']
eight_blocks_responses = pd.DataFrame(columns=['negative bias', 'unbias'])
eight_blocks_responses['negative bias'] = biased_responses['8 Blocks']
eight_blocks_responses['unbias'] = unbias_responses['8 Blocks']
```

```python
passing_period_responses.head()
```

```python
sns.set_theme(style="ticks")
ax = sns.boxplot(data=passing_period_responses, orient="h", )
ax.set(xlabel='Response (rating between 1 and 10)', ylabel='Question Type', title='Passing Period Question')
```

```python
ax = sns.boxplot(data=eight_blocks_responses, orient="h")
ax.set(xlabel='Response (rating between 1 and 10)', ylabel='Question Type', title='8 Blocks Question')
```

```python
# histogram comparisons of passing periods
biased_passing_period = biased_responses['Passing Period']
unbias_passing_period = unbias_responses['Passing Period']
plt.hist(biased_passing_period, alpha=0.5, label='biased')
plt.hist(unbias_passing_period, alpha=0.5, label='unbias')
plt.legend(loc='upper right')
plt.title('Passing Period Question')
plt.xlabel('Response (rating between 1 and 10)')
plt.ylabel('Frequency')
plt.show()

```

```python
# same for 8 blocks
biased_eight_blocks = biased_responses['8 Blocks']
unbias_eight_blocks = unbias_responses['8 Blocks']
plt.hist(biased_eight_blocks, alpha=0.5, label='biased')
plt.hist(unbias_eight_blocks, alpha=0.5, label='unbias')
plt.legend(loc='upper right')
plt.title('8 Blocks Question')
plt.xlabel('Response (rating between 1 and 10)')
plt.ylabel('Frequency')
plt.show()
```

```python
# fancy 3d chart  to compare all bias
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = biased_responses['Passing Period']
y = biased_responses['8 Blocks']
z = biased_responses['8 Blocks']
ax.scatter(x, y, z, c='r', marker='o')

```
